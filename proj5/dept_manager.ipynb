```python
# 부서 정보를 저장하는 클래스 정의
# 부서 정보를 출력하는 메서드
class Dept : 
    def __init__(self, deptno, dname, loc):
        self.deptno = deptno
        self.dname = dname
        self.loc = loc

    def __str__(self) :
        return '{}\t{}\t{}'.format(self.deptno, self.dname, self.loc)
```

```python
# 기능 클래스

class DeptController:
    def __init__(self):
        self.list_dept = []

    # 부서정보 입력 메서드
    def add_dept(self, new_dept) :
        self.list_dept.append(new_dept)
        print(f'{new_dept.deptno}부서정보가 입력되었습니다.')

    # 사용자로 부터 부서 정보를 받는 메서드
    def input_dept(self):
        input_data = input('부서정보를 입력해주세요.\n 예) 10,개발팀,서울')
        dept_data = input_data.split(',')
        # print(dept_data)
        print('--')

        return Dept(dept_data[0], dept_data[1], dept_data[2])
    
    # 전체 부서 리스트를 출력하는 메서드
    def print_list(self) :
        print('번호\t이름\t위치')
        print('-----------------')
        for d in self.list_dept :
            print(d)
        
        print('-- 부서 정보 전체 리스트 출력 완료')

    # 부서 번호를 기반으로 index 값을 찾는 메서드
    def search_index(self, deptno) :
        index = -1
        for idx, d in enumerate(self.list_dept) :
            # print('>>>>>>>>>>>> ', idx, d)
            if d.deptno == deptno :
                # print('s_index : ' , idx)
                return idx
        
        return index

    # 부서번호로 정보 삭제
    def delete_dept(self) :
        deptno = input('삭제할 부서번호를 입력하세요. >>>')
        # 부서번호로 index 위치 찾기
        search_idx = self.search_index(deptno)

        #print('search_idx', search_idx)

        if search_idx > -1 :
            del self.list_dept[search_idx]
            print(deptno, '번 부서정보가 삭제되었습니다.')
        else :
            print('찾으시는 부서정보가 존재하지 않습니다.')
        
        print('-- 삭제 기능 종료')
            

    # 부서번호로 정보 검색
    def search_dept(self) :
        deptno = input('검색할 부서번호를 입력하세요. >>>')
        # 부서번호로 index 위치 찾기
        search_idx = self.search_index(deptno)

        #print('search_idx', search_idx)

        if search_idx > -1 :            
            print(deptno, '번 부서정보 ----------')
            print(self.list_dept[search_idx])
        else :
            print('찾으시는 부서정보가 존재하지 않습니다.')
        
        print('-- 검색 기능 종료')

    # 부서번호로 검색 후 정보 수정
    def edit_dept(self) :
        deptno = input('수정할 부서의 부서번호를 입력하세요. >>>')
        # 부서번호로 index 위치 찾기
        search_idx = self.search_index(deptno)        

        if search_idx > -1 :
            self.list_dept[search_idx] = self.input_dept()
            print('정보가 수정되었습니다.')
        else :
            print('찾으시는 부서정보가 존재하지 않습니다.')
            
        print('-- 수정 기능 종료')
            
```

```python
# main flow

menu_display = """-------------------------------------------------------------------------------------------------------------------------
# 메뉴구성 :  #1. 전체리스트출력  #2. 부서정보검색  #3. 부서정보입력  #4. 부서정보수정  #5. 부서정보삭제  #6. 프로그램종료(EXIT)
-------------------------------------------------------------------------------------------------------------------------"""

controller = DeptController()

while True:
    print(menu_display)
    print('메뉴 입력하세요')
    select_nenu = int(input('메뉴 번호 입력'))
    

    match select_nenu:
        case 1 :            
            controller.print_list()
        case 2 :
            controller.search_dept()
        case 3 :
            controller.add_dept(controller.input_dept())            
        case 4 :
            controller.edit_dept()
        case 5 :
            controller.delete_dept()
        case 6 :
            print('프로그램을 종료합니다.')
            break
        case _ :
            print('정확한 메뉴를 입력해주세요')



```
